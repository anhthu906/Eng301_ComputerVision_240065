```python
import cv2

image = cv2.imread('input.bmp')              # numpy array
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray, (7,7), 0)

cv2.imwrite('output_grayscale.bmp', blur)
print('Saved as output_grayscale.bmp')
```

```python
from PIL import Image
import cv2

image = Image.open('input.bmp')
pixels = image.load()
width, height = image.size
gray_image = Image.new('L', (width, height))
gray_pixels = gray_image.load()

for y in range(height):
    for x in range(width):
        r, g, b = pixels[x, y]  
        #code chỉ nhận hệ 3 màu, không nhận nhiều hơn
        #ví dụ như là RGBA trong đó Alpha là hệ màu chồng lên nhau chẳng hạn thì 
        #nó đang là 4 hệ màu nên 3 biến không detect được

        gray = int(0.299*r + 0.587*g + 0.114*b)
        gray_pixels[x, y] = gray

gray_image.save('output_grayscalePIL.bmp')
print('Saved as output_grayscale.bmp')
      
```
